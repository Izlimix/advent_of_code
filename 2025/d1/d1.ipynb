```python
filename = "sample.txt"
# filename = "input.txt"
with open(filename, encoding="utf-8") as f:
    data = f.read()

lines = data.strip().split("\n")
```

https://adventofcode.com/2025/day/1

```python
directions = {
    "R": +1,
    "L": -1
}
```

```python
# Part 1
current = 50
password = 0
for l in lines:
    direction = l[0]
    distance = int(l[1:])
    move = directions[direction] * distance
    current += move
    current = current % 100
    if current == 0:
        password += 1

print(password)
```

```python
## Part 2
def get_password2(lines: list[str], *, verbose: bool = False) -> int:
    current = 50
    password = 0
    for l in lines:
        direction = l[0]
        distance = int(l[1:])
        full_rotations, actual_distance = divmod(distance, 100)
        move = directions[direction] * actual_distance
        start = current
        end = start + move
        current = end % 100

        # password
        # Partial turn reaching 0
        partial_rotations = 0
        if start != 0:
            if current == 0:
                # Landed on 0
                partial_rotations = 1
            elif end != current:
                # Passed 0
                partial_rotations = 1

        password += full_rotations + partial_rotations
        if verbose:
            print(f"{start=} {l=} {current=}, {full_rotations=}, {partial_rotations=}, {password=}")
    return password

print(get_password2(lines))
```

```python
"""
Edge cases
- Land on 0
- Start on 0, then make a partial turn (L or R) => password += 0
- Start on 1, L2, end on 99 => password += 1
- Start on 1, R199, end on 0 => password += 2
"""
edge_cases = [
    "L50",
    "R200",
    "R1000",
    "L1",
    "R2"
]
get_password2(edge_cases, verbose=True)
```
