```python
from collections import deque
from tqdm import tqdm
```

```python
filename = "sample.txt"
# filename = "input.txt"
with open(filename, encoding="utf-8") as f:
    data = f.read()

lines = data.strip().split("\n")
```

```python
height = len(lines)
width = len(lines[0])
splitters = set()
start = -1
for y, line in enumerate(lines):
    for x, c in enumerate(line):
        pos = complex(x, y)
        match c:
            case "S":
                start = pos
            case "^":
                splitters.add(pos)
            case _:
                pass

if start == -1:
    raise NotImplementedError("Missing start")
```

```python
## Part 1
seen = set()
pending = deque([start])
while pending:
    pos = pending.popleft()
    if pos in seen:
        continue
    # Out of bounds
    if (not (0 <= pos.real < width)) or (not (0 <= pos.imag < height)):
        continue
    seen.add(pos)
    if pos in splitters:
        pending.extend([pos - 1 + 1j, pos + 1 + 1j])
    else:
        pending.append(pos + 1j)

len(seen & splitters)
```

```python
## Part 2
seen = set()
pos_timelines = {start: 1}
total = 0
pending = deque([start])
while pending:
    pos = pending.popleft()
    if pos in seen:
        continue
    seen.add(pos)
    timelines = pos_timelines[pos]
    # Out of bounds
    if (not (0 <= pos.real < width)) or (not (0 <= pos.imag < height)):
        total += timelines
        continue
    # Step
    if pos in splitters:
        nexts = [pos - 1 + 1j, pos + 1 + 1j]
    else:
        nexts = [pos + 1j]

    for next_pos in nexts:
        pending.append(next_pos)
        pos_timelines[next_pos] = pos_timelines.get(next_pos, 0) + timelines

total
```
