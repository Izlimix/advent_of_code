```python
from operator import itemgetter
```

```python
filename = "sample.txt"
# filename = "input.txt"
with open(filename, encoding="utf-8") as f:
    data = f.read()

lines = data.strip().split("\n")
```

```python
## Part 1
part1 = 0
for l in lines:
    # Largest first digit can't be the last character (Assuming len > 1)
    l = list(enumerate(l))
    ind1, v1 = max(l[:-1], key=itemgetter(1))
    ind2, v2 = max(l[ind1 + 1:], key=itemgetter(1))
    joltage = int(v1 + v2)
    # print(f"{joltage=}")
    part1 += joltage

part1
```

```python
# Part 2
# Turn on 12 batteries per bank instead
# Is caching enough for this?
# -> There are 1_050_421_051_106_700 (1e15) combinations of C(100, 12). So probably not.
# Greedy algorithm. Just make sure to always pick digits with at least (12 - ind) chars left
part2 = 0
max_batteries = 12
for l in lines:
    l = list(enumerate(l))
    start = 0
    joltage = ""
    for next_digits in reversed(range(0, max_batteries)):
        if next_digits == 0:
            candidates = l[start:]
        else:
            candidates = l[start:-next_digits]
        ind, v = max(candidates, key=itemgetter(1))
        joltage += v
        start = ind + 1
    # print(f"{joltage=}")
    part2 += int(joltage)
    
part2
```
