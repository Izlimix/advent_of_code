```python
from tqdm import tqdm
```

```python
filename = "sample.txt"
# filename = "input.txt"
with open(filename, encoding="utf-8") as f:
    data = f.read()

raw_ranges, raw_ids = data.strip().split("\n\n")
```

```python
range_pairs = []
ranges = []
for l in raw_ranges.split("\n"):
    a, b = map(int, l.split("-"))
    range_pairs.append((a, b))
    ranges.append(range(a, b + 1))
ids = list(map(int, raw_ids.split("\n")))
```

```python
## Part 1
fresh_ids = [x for x in ids if any(x in r for r in ranges)]
len(fresh_ids)
```

```python
## Part 2
# def ranges_overlap(x_lo, x_hi, y_lo, y_hi) -> bool:
#     return (x_lo <= y_hi) and (y_lo <= x_hi)

def ranges_overlap(x, y) -> bool:
    return (x[0] <= y[1]) and (y[0] <= x[1])

# def combine_ranges(x_lo, x_hi, y_lo, y_hi) -> tuple:
#     return (min(x_lo, y_lo), max(x_hi, y_hi))

def combine_ranges_multiple(pairs: list[tuple[int, int]]) -> tuple[int, int]:
    lows, highs = zip(*pairs)
    return (min(lows), max(highs))
```

```python
merged_ranges = {range_pairs[0]}
for r1 in tqdm(range_pairs[1:]):
    overlaps = [r0 for r0 in merged_ranges if ranges_overlap(r0, r1)]
    r1_merged = combine_ranges_multiple([r1, *overlaps])
    merged_ranges.difference_update(overlaps)
    merged_ranges.add(r1_merged)

sum(hi - lo + 1 for lo, hi in merged_ranges)
```
