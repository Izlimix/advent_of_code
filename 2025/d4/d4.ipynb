```python
from tqdm import tqdm
```

```python
filename = "sample.txt"
# filename = "input.txt"
with open(filename, encoding="utf-8") as f:
    data = f.read()

lines = data.strip().split("\n")
```

```python
rolls = set()
for y, line in enumerate(lines):
    for x, c in enumerate(line):
        pos = complex(x, y)
        if c == "@":
            rolls.add(pos)
```

```python
# Part 1
adjacents = {complex(x, y) for x in (1, 0, -1) for y in (1, 0, -1)} - {0}
def find_accessible_rolls(rolls: set[complex]) -> set[complex]:
    accessible_rolls = set()
    for roll in rolls:
        neighbours = {roll + step for step in adjacents}
        if len(rolls & neighbours) < 4:
            accessible_rolls.add(roll)
    return accessible_rolls

accessible_rolls = find_accessible_rolls(rolls)
len(accessible_rolls)
```

```python
# Part 2
current = set(rolls)
with tqdm(total=len(current)) as pbar:
    while accessible_rolls:
        current -= accessible_rolls
        accessible_rolls = find_accessible_rolls(current)
        pbar.update()

len(rolls - current)
```
