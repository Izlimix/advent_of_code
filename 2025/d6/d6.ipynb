```python
import math
```

```python
filename = "sample.txt"
# filename = "input.txt"
with open(filename, encoding="utf-8") as f:
    data = f.read()

lines = data.split("\n")
```

```python
vs = [l.split() for l in lines]
cols = list(zip(*vs, strict=True))
# cols
```

```python
## Part 1
answers = []
for *ds, op in cols:
    ds = map(int, ds)
    match op:
        case "+":
            ans = sum(ds)
        case "*":
            ans = math.prod(ds)
        case _:
            raise NotImplementedError(op)
    answers.append(ans)
sum(answers)
```

```python
## Part 2
# Need to re-parse blocks, preserving extra spaces since ds can be left- or right-justified
raw_cols = list(zip(*lines, strict=True))
# raw_cols
```

```python
answers = []
block = []
# Read right-to-left. Operators are in the leftmost col of the block
for *ds, op in reversed(raw_cols):
    if all(d == " " for d in ds):
        continue
    block.append(int("".join(ds)))
    match op:
        case "+":
            ans = sum(block)
        case "*":
            ans = math.prod(block)
        case _:
            continue
    answers.append(ans)
    block = []
sum(answers)
```
